```python
def make_df(cols, ind):
    """Stwórz DataFrame z kombinacji cols i ind"""
    data = {c: [str(c) + str(i) for i in ind]
            for c in cols}
    return pd.DataFrame(data, ind) 
```

## `.apply()` z `lambda`

Funkcja `.apply()` w Pandas pozwala na wykonywanie operacji na wierszach lub kolumnach DataFrame, gdzie logika obliczeń może być bardziej skomplikowana. Najczęściej używana z `lambda`, sprawdza się w sytuacjach, gdy potrzebujemy:

1. Zastosować niestandardowe obliczenia na poziomie wiersza lub kolumny.
2. Wykorzystać warunki lub złożoną logikę, np. różne wzory w zależności od wartości w jednej lub kilku kolumnach.



### Przykład:
Chcemy obliczyć nową kolumnę, w której:
- Jeśli wartość w kolumnie A jest większa niż w kolumnie B, wynik to `A^2 - 2 * B`.
- W przeciwnym razie wynik to `2 * B^2 + A`.

W takim przypadku `.apply()` z `lambda` pozwala na prostą i czytelną implementację tej logiki.


```python
import pandas as pd

df = pd.DataFrame({
    'A': [1, 4, 3, 7],
    'B': [2, 2, 3, 6]
})

df['Wynik'] = df.apply(lambda row:
                           row['A']**2 - 2*row['B']
                           if row['A'] > row['B']
                           else 2*row['B']**2 + row['A'],
                       axis=1)

print(df)
```

## Operacje wektorowe

Operacje wektorowe w Pandas opierają się na bezpośrednich obliczeniach na całych kolumnach bez iteracji po wierszach. Są one:
- Szybsze, ponieważ korzystają z niskopoziomowych optymalizacji w NumPy.
- Bardziej wydajne w przypadku prostych operacji matematycznych, takich jak dodawanie, mnożenie czy zastosowanie jednego warunku.

### Przykład:
Chcemy obliczyć nową kolumnę, która zawsze jest równa `0.7 * A + 0.3 * B`. W takim przypadku możemy bez problemu wykonać operację wektorową na kolumnach A i B.


```python
df['Wynik_wektorowo'] = 0.7 * df['A'] + 0.3 * df['B']

print(df)
```
